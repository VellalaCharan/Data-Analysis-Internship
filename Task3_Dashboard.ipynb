```python
# Import required libraries
import pandas as pd              # For data handling
import matplotlib.pyplot as plt  # For visualization

# Upload dataset from your system (Google Colab)
from google.colab import files
uploaded = files.upload()

# Load dataset into pandas DataFrame
data = pd.read_csv("netflix_titles.csv")

# Display first 5 rows to understand data
data.head()
```

```python
# 📊 1. Movies vs TV Shows Count

# Count number of each type (Movie / TV Show)
type_counts = data['type'].value_counts()

# Plot bar chart
type_counts.plot(kind='bar')

# Add title and labels
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")

# Show graph
plt.show()
```

```python
# 📊 2. Top Release Years Analysis

# Count number of shows per year (top 10 years)
year_counts = data['release_year'].value_counts().head(10)

# Plot bar chart
year_counts.plot(kind='bar')

# Add title and labels
plt.title("Top Release Years")
plt.xlabel("Year")
plt.ylabel("Count")

# Show graph
plt.show()
```

```python
# 📊 3. Top Countries Analysis

# Count shows by country (top 10 countries)
country_counts = data['country'].value_counts().head(10)

# Plot bar chart
country_counts.plot(kind='bar')

# Add title and labels
plt.title("Top Countries Producing Content")
plt.xlabel("Country")
plt.ylabel("Count")

# Show graph
plt.show()
```
