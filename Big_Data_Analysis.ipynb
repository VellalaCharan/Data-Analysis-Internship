```python
# Import required libraries
# PySpark is used for big data processing
from pyspark.sql import SparkSession

# Create Spark session (this starts the Spark engine)
spark = SparkSession.builder.appName("BigDataAnalysis").getOrCreate()

# Load the CSV dataset
# header=True means first row is column names
# inferSchema=True automatically detects data types
data = spark.read.csv("netflix_titles.csv", header=True, inferSchema=True)

# Show first 10 rows of the dataset
print("First 10 rows of dataset:")
data.show(10)

# Print dataset structure (columns and data types)
print("Dataset Schema:")
data.printSchema()

# Count total number of records in dataset
print("Total records in dataset:")
print(data.count())

# Analysis 1: Count number of Movies vs TV Shows
print("Movies vs TV Shows count:")
data.groupBy("type").count().show()

# Analysis 2: Count shows by country
print("Top countries producing content:")
data.groupBy("country").count().orderBy("count", ascending=False).show()

# Analysis 3: Find latest released shows
print("Latest released content:")
data.orderBy("release_year", ascending=False).show(5)
```
